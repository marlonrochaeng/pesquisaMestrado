```python
import pandas as pd
```

```python
df_ag = pd.read_csv('../resultados/mesma populacao/mutacao simples/mesma_pop_ag.csv',index_col=0)
df_ag.style.format("{:.2f}")
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
df_ag.loc[df_ag['makespan'].isin(df_ag.groupby('instance')['makespan'].min().values)]
```
