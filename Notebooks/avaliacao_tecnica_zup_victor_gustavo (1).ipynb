*   Análise da variável de valor do reembolso e limite de reembolso

```python
monetary_variables = cleaned_data[['valor', 'limite_maximo']].copy()
monetary_variables.describe()
```

```python
plot_4 = monetary_variables.plot(kind = 'box', figsize = (10,8))
plot_4.set_title('Distribuição de valores e limites máximos')
```

Aqui já verificamos que alguns valores são muito discrepantes dos outros valores, inclusive da média. Isso pode indicar outliers. Vamos usar o método de intervalo interquartil para identificação e remoção desses outliers.

```python
q1 = cleaned_data['valor'].quantile(0.25)
q3 = cleaned_data['valor'].quantile(0.75)
iqr = q3 - q1

cleaned_data = cleaned_data[~((cleaned_data['valor'] < (q1 - 1.5 * iqr)) | (cleaned_data['valor'] > (q3 + 1.5 * iqr)))]
cleaned_data = cleaned_data[~((cleaned_data['limite_maximo'] < (q1 - 1.5 * iqr)) | (cleaned_data['limite_maximo'] > (q3 + 1.5 * iqr)))]
```

```python
cleaned_data.reset_index(inplace = True)

cleaned_data[['valor', 'limite_maximo']].describe()
```

```python
monetary_values_cleaned = cleaned_data[['valor', 'limite_maximo']].copy()
plot_5 = monetary_values_cleaned.plot(kind = 'box', figsize = (10, 8))
plot_5.set_title('Nova distribuição de valores e limites máximos')
```

```python
plot_6 = cleaned_data[['limite_maximo', 'valor']].plot(kind = 'hist', figsize = (12, 6), grid = True, alpha = 0.6, bins = 20)
plot_6.set_title('Frequência de valores de reembolso')
plot_6.set_ylabel('Frequência')
plot_6.set_xlabel('Valor')
plot_6.autoscale()
```

Aqui já podemos verificar a discrepância de valores entre os valores solicitados e os limites máximos. A seguir, vamos começar a criar algumas inferências com relação aos dados obtidos.

*    Teste de hipótese de correlação entre cargos e categorias de solicitação

```python
cross_res = pd.crosstab(index = cleaned_data['ds_funcao'], columns = cleaned_data['categoria'])
chi2_contingency(cross_res)[1]
```

Tendo H0: As variáveis não possuem correlação, com o resultado do teste Qi² acima sendo menor que 0,5, podemos descartar a hipótese nula, e dizer que há indícios suficientes para afirmar que as variáveis estão correlacionadas. Assim, podemos analisá-las graficamente. Agora, façamos o mesmo teste para cargos e a sub-categoria.

```python
cross_res = pd.crosstab(index = cleaned_data['ds_funcao'], columns = cleaned_data['sub_categoria'])
chi2_contingency(cross_res)[1]
```

Assim, para o mesmo teste, verificamos que também há indícios suficientes para afrimar que essas variáveis possuem correlação entre si. Façamos então o cruzamento dessas variáveis e sua análise.

### Conclusões

O cálculo de limite máximo de solicitações de reembolso é um fator importante a se considerar para uma boa saúde financeira de uma empresa. Este cálculo, quando feito de forma incorreta, pode gerar problemas como falta de caixa, gastos não planejados, entre outros fatores. Após o tratamento e a análise dos dados fornecidos para esta situação. Podem-se elencar alguns pontos que podem ser importantes de se observar para o ajuste desses limites. Dentre alguns, elencam-se:



*   Zuppers da área de desenvolvimento requisitam mais reembolsos que outras funções;
*   Reembolsos relacionados a alimentação e treinamento são os mais frequentes no conjunto de dados;
*   A solicitação de valores, para a maioria dos valores, é distribuída de forma normal, apesar de haverem valores que extrapolem essas condições;
*   As solicitações parecem seguir um padrão em que em um período do ano (de Agosto a Outubro) são menos frequentes, enquanto que nos outros períodos parecem se manter numa frequência semelhante, embora os dados trabalhados não apresentem mais de um ano para se confirmar o padrão;
*   A categoria de reembolso por hora extra possui uma maior frequência de solicitação, enquanto a sub-categoria de alimentação ocupa essa posição;
*   A função `Business P&D - 2`, apesar de não ser a que solicita reembolso com mais frequência, é a que solicita os maiores valores;
*   As funções `Developer` - 1 e `Developer - 3` solicitam valores que estão, em média, adequados com os limites máximos;
*   Um único zupper requisita os maiores valores e com mais frequência, exercendo ele a função `Business P&D - 2`;
*   Os valores solicitados, por categorias e sub-categorias, tendem a ser maiores que os limites máximos para estas.

Assim, de posse destes indícios, e com a combinação dos mesmos, podem ser propostas algumas ideias para a melhoria desses limites, a saber:


*   Considerar a função dos zuppers e, a partir de conjuntos de funções, aumentar o limite máximo de reembolso;
*   Zuppers com funções administrativas, que viajam muito ou que fazem muita hora extra podem ter os limites máximos de suas solicitações aumentados;
*   Reembolsos relacionados a horas extras e alimentação, de acordo com cada cenário, podem ter os limites máximos aumentados;
*   Em determinadas épocas do ano, podem ser feitas revisões dos limites máximos para os tipos de solicitações diferentes, aumentando ou diminuindo os mesmos com base nos parâmetros apresentados;
*   Por fim, a criação de um modelo, seja ele utilizando regressão linear, métodos de inteligência artificial mais simples ou mais complexos, para estimar o limite máximo com base nestas informações também pode ser interessante, pois diminuiria a interação humana nesses cenários, diminuindo a necessidade de ajustes manuais, agilizando o processo, e podendo aumentar a precisão desses reembolsos, gerando assim mais satisfação para os zuppers e mais economia para a própria empresa.
