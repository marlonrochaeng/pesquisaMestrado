```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
df = pd.read_csv('eda_com_heuristicas_e_mut.csv',index_col=0)
#df.style.format("{:.2f}")
df.head()
```

```python
df_min = df.loc[df['makespan'].isin(df.groupby('instance')['makespan'].min().values)]
df_min.sort_values('instance',inplace=True, ascending=True)
```

```python
df_min
```

```python
df_u_c_hihi = pd.read_csv('eda_populacao_gerada_paulo.csv',index_col=0)
df_a = df_u_c_hihi.loc[df_u_c_hihi['makespan'].isin(df_u_c_hihi.groupby('instance')['makespan'].min().values)]
df_a.sort_values('instance',inplace=True, ascending=True)
#df_min.sort_values('instance',inplace=True, ascending=True)
```

```python
df_a = df_a.drop_duplicates(subset = ["makespan"])
```

```python
df_a
```

```python
df_min = df_min.drop(columns=['jobs', 'machines','exec_time','selection_method'])
df_a = df_a.drop(columns=['jobs', 'machines','exec_time','selection_method'])
df_ags = pd.concat([df_min.reset_index(), df_a.reset_index()], 
                   axis='columns', keys=['Sem pop. inicial', 'Com pop. inicial'])
df_ags['improvement'] = 1 - df_ags['Com pop. inicial']['makespan']/df_ags['Sem pop. inicial']['makespan']
df_ags
```

```python
df_mutacao_correto = pd.read_csv('populacao_grande_mutacao.csv',index_col=0)
df_a_correto = df_mutacao_correto.loc[df_mutacao_correto['makespan'].isin(df_mutacao_correto.groupby('instance')['makespan'].min().values)]
df_a_correto.sort_values('instance',inplace=True, ascending=True)
```

```python
df_a_correto.drop_duplicates(subset=['makespan'])
```

```python
df_to_compare = pd.DataFrame(columns=['Algorithm', 'Instance', 'Passamos', 'Quanto falta','Makespan do algoritmo', 'Nosso makespan'])
df_to_compare['Instance'] = ['u_c_hihi.0', 'u_c_hilo.0', 'u_c_lohi.0', 'u_c_lolo.0','u_i_hihi.0','u_i_hilo.0',
    'u_i_lohi.0','u_i_lolo.0','u_s_hihi.0', 'u_s_hilo.0', 'u_s_lohi.0', 'u_s_lolo.0']
df_to_compare['Algorithm'] = 'pCHC - average'
df_to_compare['Nosso makespan'] = list(df_a_correto.drop_duplicates(subset=['makespan'])['makespan'])
df_to_compare['Makespan do algoritmo'] = [7481195,153924,243446.3,5181.6, 2956906,73847.1,102677.3,2557.2,4239146,96750.3,123989.4,3472.2]
df_to_compare.loc[df_to_compare['Makespan do algoritmo'] > df_to_compare['Nosso makespan'], 'Passamos'] = 'True' 
df_to_compare.loc[df_to_compare['Makespan do algoritmo'] < df_to_compare['Nosso makespan'], 'Passamos'] = 'False' 
df_to_compare['Quanto falta'] =  df_to_compare['Nosso makespan'] - df_to_compare['Makespan do algoritmo']
```

```python
df_to_compare
```
