```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
df = pd.read_csv('eda.csv',index_col=0)
#df.style.format("{:.2f}")
df.head()
```

```python
df_min = df.loc[df['makespan'].isin(df.groupby('instance')['makespan'].min().values)]
df_min.sort_values('instance',inplace=True, ascending=True)
```

```python
df_min
```

```python
df_u_c_hihi = pd.read_csv('eda_uc_hihi.csv',index_col=0)
df_a = df_u_c_hihi.loc[df_u_c_hihi['makespan'].isin(df_u_c_hihi.groupby('instance')['makespan'].min().values)]
df_a.sort_values('instance',inplace=True, ascending=True)
#df_min.sort_values('instance',inplace=True, ascending=True)
```

```python
df_a = df_a.drop_duplicates(subset = ["makespan"])
```

```python
df_a
```

```python
df_min = df_min.drop(columns=['jobs', 'machines','exec_time','selection_method'])
df_a = df_a.drop(columns=['jobs', 'machines','exec_time','selection_method'])
df_ags = pd.concat([df_min.reset_index(), df_a.reset_index()], 
                   axis='columns', keys=['Sem pop. inicial', 'Com pop. inicial'])
df_ags['improvement'] = 1 - df_ags['Com pop. inicial']['makespan']/df_ags['Sem pop. inicial']['makespan']
df_ags
```
